# Preparing for Modelling

#### Further Data Exploration for Modelling

* Autocorrelation plot

```python
plt.rc("figure", figsize=(10,6))
pd.plotting.autocorrelation_plot(data['Wheat'])
```

### Importing more libraries for Model Building

```python
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
```

```python
plt.rc("figure", figsize=(10,6))
sm.graphics.tsa.plot_acf(data['Wheat'], lags = 275);
```

* A closer look

```python
plt.rc("figure", figsize=(10,6))
sm.graphics.tsa.plot_acf(data['Wheat'], lags = 80);
```

* Partial Autocorrelation

```python
plt.rc("figure", figsize=(10,6))
sm.graphics.tsa.plot_pacf(data['Wheat']);
```

Is the data stationay?

```python
X = data['Wheat'].values

result = adfuller(X)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Lags used: %f' % result[2])
print('Critical Values:')
for key, value in result[4].items():
	print('\t%s: %.3f' % (key, value))
```

Because p-value of 0.112468 > 0.05, we fail to reject the null. Therefore, the data is non-stationary

### Making the Data Stationary

##### Removing the Trend

* ACF

```python
data_diff = data['Wheat'] - data['Wheat'].shift(1)
plt.rc("figure", figsize=(10,6))
sm.graphics.tsa.plot_acf(data_diff[1:], lags=40);
```

```python
plt.figure(figsize=(30,8))
sns.lineplot(data=data_diff[1:], palette="deep");
```

* PACF

```python
plt.rc("figure", figsize=(10,6))
sm.graphics.tsa.plot_pacf(data_diff);
```

```python
X = data_diff.values

result = adfuller(X)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Lags used: %f' % result[2])
print('Critical Values:')
for key, value in result[4].items():
	print('\t%s: %.3f' % (key, value))
```

Because p-value of 0 < 0.05, we reject the null. Therefore, the data is stationary

## Modelling

### Using AutoARIMA

```python
data_diff
```

```python
data_diff = data_diff.dropna()
```

```python
#divide into train and validation set
train = data_diff[:int(0.7*(len(data)))]
test = data_diff[int(0.7*(len(data))):]

#plotting the data
train.plot()
test.plot()
```

```python
#!pip install pmdarima
```

```python
#building the model
from pmdarima.arima import auto_arima
model = auto_arima(train, trace=True, error_action='ignore', suppress_warnings=True)
model.fit(train)

forecast = model.predict(n_periods=len(test))
forecast = pd.DataFrame(forecast,index = test.index,columns=['Prediction'])

#plot the predictions for validation set
plt.plot(train, label='Train')
plt.plot(test, label='Test')
plt.plot(forecast, label='Prediction')
plt.show()
```

```python
#calculate rmse
from math import sqrt
from sklearn.metrics import mean_squared_error

rmse = sqrt(mean_squared_error(test,forecast))
print(rmse)
```

### Rolling Average

```python
def ma(df, coloumn='Wheat'):
    df['MA'] = df[coloumn].rolling(window=3).mean()
    return df
```

```python
data1 = ma(data)
data1
```

```python
plt.figure(figsize=(26,8))
plt.title('Variation of wheat along with moving averages. ', fontsize=18)
plt.plot(data1.Wheat)
plt.plot(data1.MA)
plt.ylabel('Price', fontsize=18)
plt.xlabel('Index', fontsize=18)
plt.legend(['close_price','moving Average'], loc='upper left');
```

### Using SARIMAX with no exogenous variables

```python
#divide into train and validation set
train = data[:int(0.7*(len(data)))]
test = data[int(0.7*(len(data))):]

#plotting the data
train['Wheat'].plot()
test['Wheat'].plot()
```

```python
train
```

```python
test
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
my_order = (1,0,0)
my_seasonal_order = (0,0,0,0)
model = SARIMAX(train['Wheat'], order = my_order, seasonal_order=my_seasonal_order).fit()
```

```python
predictions = model.forecast(len(test))
residuals = test['Wheat'] - predictions
plt.plot(residuals)
```

```python
predictions_df = pd.DataFrame(predictions)
predictions_df["Date"] = test.index
```

```python
plt.figure(figsize=(20,20))
plt.plot(data['Wheat'])
plt.plot(predictions_df["Date"], predictions_df["predicted_mean"])
```

```python
#RMSE
print(np.sqrt(np.mean(residuals**2)))
```

### Using SARIMAX with exogenous variables

```python
data.isna().sum()
```

```python
data = data.dropna()
data.info()
```

```python
train = data[:int(0.7*(len(data)))]
test = data[int(0.7*(len(data))):]
```

```python
train
```

```python
my_order = (1,0,0)
my_seasonal_order = (0,0,0,0)
model = SARIMAX(train['Wheat'], exog = train[['Food Price Index', 'Food and Beverage Price Index', 'Maize corn', 
                                              'Soybeans', 'Non-Fuel Price Index', 'Soybean Oil', 'Barley', 
                                              'Rapeseed oil', 'All Commodity Price Index', 'Soybean Meal', 'Palm oil',
                                              'Coal', 'Fuel Energy Index', 'Lead', 
                                              'Crude Oil - petroleum - Dated Brent light blend', 'Oil Dubai', 
                                              'Crude Oil petroleum', 'Crude_Oil_Avg', 'Tin']], order = my_order, 
                seasonal_order = my_seasonal_order).fit()
```

```python
model.summary()
```

```python
predictions = model.forecast(len(test), exog=test[['Food Price Index', 'Food and Beverage Price Index', 
                                                         'Maize corn', 
                                              'Soybeans', 'Non-Fuel Price Index', 'Soybean Oil', 'Barley', 
                                              'Rapeseed oil', 'All Commodity Price Index', 'Soybean Meal', 'Palm oil',
                                              'Coal', 'Fuel Energy Index', 'Lead', 
                                              'Crude Oil - petroleum - Dated Brent light blend', 'Oil Dubai', 
                                              'Crude Oil petroleum', 'Crude_Oil_Avg', 'Tin']])
residuals = test['Wheat'] - predictions
plt.plot(residuals)
```

```python
predictions_df = pd.DataFrame(predictions)
predictions_df["Date"] = test.index
```

```python
plt.figure(figsize=(20,20))
plt.plot(data.index, data['Wheat'])
plt.plot(predictions_df["Date"], predictions_df["predicted_mean"])
```

```python
#RMSE
print(np.sqrt(np.mean(residuals**2)))
```
